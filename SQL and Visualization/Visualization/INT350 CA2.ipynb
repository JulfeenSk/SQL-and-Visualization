```python
#Importing the necessary libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

```

```python
# loading the dataset as pandas dataframe
data = pd.read_csv("data.csv")
data

```

```python
# To analyse columns in the dataset 
data.info()
```

```python
# creating boxplot
fig, axes = plt.subplots(1, 2, figsize=(10, 4)) 

axes[0].boxplot(data['ApplicantIncome'])
axes[0].set_title("Boxplot of ApplicantIncome")

axes[1].boxplot(data['CoapplicantIncome'])
axes[1].set_title("Boxplot of CoapplicantIncome")

plt.show()
```

**Interpretation of the Boxplot**

- The box in the middle represents the interquartile range (IQR), which contains the middle 50% of the data.
- The line inside the box represents the median of the data.
- The lines(whiskers) extend from the box to the smallest and largest values within 1.5 times the IQR. Data points beyond the lines are considered potential outliers.
- Individual data points outside the lines are shown as individual points and are often considered outliers.
- Outliers can be values that are significantly higher or lower than the majority of the data.
- The boxplot helps to visually identify the spread and distribution of data, as well as the presence of outliers.

6.Briefly discuss about Stacked Bar Graphs with an example

**Stacked Bar Graphs**

- A stacked bar graph is a type of bar chart where multiple data series are stacked on top of each other, rather than being shown side by side.
- It's used to visualize the composition of a whole by showing the contributions of different categories or subgroups to that whole.
- Each bar in a stacked bar graph represents the total quantity of a variable, and the segments within the bar represent how that quantity is divided among different categories or subgroups 
- A stacked bar graph is a type of data visualization used to represent the composition of a whole by showing how different categories or subgroups contribute to that whole
- The height of each segment within a bar indicates its contribution to the total

**Example in context with above dataset**

```python
import numpy as np

pivot_df = df.pivot_table(values='ApplicantIncome', index='Gender', columns='Education', aggfunc=np.sum, fill_value=0)


fig, ax = plt.subplots(figsize=(10, 10))
pivot_df.plot(kind='bar', stacked=True, ax=ax, colormap='viridis')


ax.set_xlabel('Gender')
ax.set_ylabel('ApplicantIncome')
ax.set_title('Stacked Bar Graph: LoanAmount by Education and Gender')


ax.legend(title='Education')


plt.show()
```
