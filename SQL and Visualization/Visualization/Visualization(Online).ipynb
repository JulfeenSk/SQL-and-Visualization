### SQL in jupyter notebook

```python
# import sqlite3 module
import sqlite3 
```

```python
# while using sqlite3 we need a cinnection and a cursor
# creting a connetion object -- most ddl commonad srae excecuted  
# creating a new database student_marks

connection = sqlite3.connect('student_marks.db')
```

```python
# we can excecute anything using cursor
# most dml commands are used using cursor

cursor = connection.cursor()
print('Intialized the DB')
```

```python
# write a sql query
query  = 'Select sqlite_version()'
cursor.execute(query)

print(cursor.fetchall())
```

```python
# creat a table sql_marks

connection.execute('''
CREATE TABLE sql_marks
    (S_ID INT PTIMARY KEY,
     S_NAME TEXT NOT NULL,
     MARKS INT NOT NULL 
    );
''')
```

```python
# insertion of data into this table

connection.execute("INSERt INTO sql_marks VALUES (1001,'AMAN',85),(1002,'AKASH',65)")
```

```python
connection.execute("INSERt INTO sql_marks VALUES (1003,'AJAY',95)")
```

```python
# Fetch the details

cursor = connection.execute("SELECT * FROM sql_marks")
```

```python
print(cursor.fetchall())
```

```python
# we need to assign value to cursor everytime before excecuting
for row in cursor:
    print(row)
```

```python
connection.close()
```

```python
import seaborn as sns 
import matplotlib as plt 
 
```

```python
# importing module penguins from seaborn
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

#### Distributional data

```python
sns.set_style('whitegrid')
```

```python
# histplot
sns.histplot(data= penguins,x='flipper_length_mm',hue='species')
```

```python
sns.histplot(data= penguins,x='flipper_length_mm',hue='species',multiple = 'stack')
```

```python
#kde - kernel density estiamtion
sns.kdeplot(data= penguins,x='flipper_length_mm',hue='species',multiple = 'stack')
```

```python
#displot - distributionoal plot  
sns.displot(data= penguins,x='flipper_length_mm',hue='species',multiple = 'stack')
# gives a result similar to histplot
# displot() is a figure-levl function for the distributions module.
#its dfeault behvious is to draw a histogram using the same code as histplot() behind the scenes 
```

### Figure - level function

```python
sns.displot(data= penguins,x='flipper_length_mm',hue='species',col = 'species')
```

```python
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde')
```

```python
# smoothening the bandwidth
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde',bw_adjust = 0.25)
```

```python
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde',bw_adjust = 2.5)
```

```python
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde',hue='species',bw_adjust = 2.5)
```

```python
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde',hue='species',bw_adjust = 0.75)
```

```python
sns.displot(data= penguins,x='flipper_length_mm',kind = 'kde',hue='species',multiple='stack',bw_adjust = 0.75)
```

```python
# kde - the data is poorly  represented 
```

#### Empherical cumulative distribution function - ECDF

```python
# ecdf plot - empherical cumulative distribution plot
sns.displot(data=penguins,x='flipper_length_mm',kind='ecdf')
```

```python
sns.displot(data=penguins,x='flipper_length_mm',hue='species',kind='ecdf')
```

```python
# bivariate 
sns.displot(data=penguins,x='bill_length_mm',y='bill_depth_mm')
```

```python
sns.displot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species')
```

```python
sns.displot(data=penguins,x='bill_length_mm',y='bill_depth_mm',kind='kde')
```

```python
sns.displot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species',kind='kde')
```
